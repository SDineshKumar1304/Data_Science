```python
import pandas as pd

# Load the data
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
tags_df = pd.read_csv('/content/tags.csv')
# Step 1: Group ratings based on movieId and apply aggregation operations
grouped_ratings = ratings_df.groupby('movieId').agg({'rating': ['count', 'mean']}).reset_index()

# Rename the columns for clarity
grouped_ratings.columns = ['movieId', 'rating_count', 'rating_mean']

# Step 2: Apply inner join on the dataframe created from movies.csv and the grouped df from step 1
merged_df = pd.merge(movies_df, grouped_ratings, on='movieId', how='inner')

# Step 3: Filter only those movies which have more than 50 user ratings
filtered_movies = merged_df[merged_df['rating_count'] > 50]

# Display the resulting DataFrame
print(filtered_movies.head())
```

```python
# Filter the DataFrame for the movie "Matrix, The (1999)"
matrix_tags = tags_df.loc[tags_df['movieId'].isin(filtered_movies[filtered_movies['title'] == 'Matrix, The (1999)']['movieId'])]

# Display the result
print("Tags submitted by users to 'Matrix, The (1999)':")
print(matrix_tags[['tag']])
```

```python
# Filter the DataFrame for Sci-Fi movies
sci_fi_movies = filtered_movies[filtered_movies['genres'].str.contains('Sci-Fi', case=False)]

# Sort the Sci-Fi movies DataFrame by rating count in descending order
sorted_sci_fi_movies = sci_fi_movies.sort_values(by='rating_count', ascending=False)

# Extract the title of the third most popular Sci-Fi movie
third_most_popular_sci_fi_movie = sorted_sci_fi_movies.iloc[2]['title']

# Display the result
print("The third most popular Sci-Fi movie based on number of user ratings is:", third_most_popular_sci_fi_movie)
```
